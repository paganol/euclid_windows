```python
import numpy as np
import matplotlib.pyplot as plt
import euclid_windows as EW
```

```python
Win = EW.Windows(bintype="equipopulated")
Win.get_distributions()
```

```python
for i in range(10):
    plt.plot(Win.zeta,Win.eta_z[i],normalize=False)
plt.plot(Win.zeta,Win.gal_dist)
```

```python
import camb
from camb import model
from camb.sources import SplinedSourceWindow
```

```python
lmax = 300

pars = camb.CAMBparams()
pars.set_cosmology(H0=67.5, ombh2=0.022, omch2=0.122)
pars.InitPower.set_params(As=2e-9, ns=0.965)
pars.set_for_lmax(lmax, lens_potential_accuracy=1)
pars.SourceTerms.counts_redshift = False
pars.SourceTerms.counts_lensing = False
pars.SourceTerms.limber_windows = True
pars.SourceTerms.limber_phi_lmin = 50
pars.SourceTerms.counts_velocity = False
pars.SourceTerms.counts_radial = False
pars.SourceTerms.counts_timedelay = False
pars.SourceTerms.counts_ISW = False
pars.SourceTerms.counts_potential = False
pars.SourceTerms.counts_evolve = True
pars.SourceTerms.line_phot_dipole = False
pars.SourceTerms.line_phot_quadrupole = False
pars.SourceTerms.line_basic = True
pars.SourceTerms.line_distortions = False
pars.SourceTerms.line_extra = False
pars.SourceTerms.line_reionization = False
pars.SourceTerms.use_21cm_mK = False

#set Want_CMB to true if you also want CMB spectra or correlations
pars.Want_CMB = True 
#NonLinear_both or NonLinear_lens will use non-linear corrections
pars.NonLinear = model.NonLinear_both
```

```python
pars.SourceWindows = [SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[0]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[1]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[2]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[3]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[4]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[5]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[6]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[7]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[8]),
                      SplinedSourceWindow(bias_z = Win.bias, z = Win.zeta, W = Win.eta_z[9]),
                      ]                          
results = camb.get_results(pars)
cls = results.get_cmb_unlensed_scalar_array_dict(CMB_unit='muK')
```

```python
ell= np.arange(2, lmax+1)
for i in np.arange(10)+1:
    plt.plot(ell,cls['TxW'+str(i)][2:lmax+1])
plt.yscale('log')
```
